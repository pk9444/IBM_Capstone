### Define some plotting and computational functions

```python
''' 
plot_hist_and_box() : a function to side by side plot a histogram and a box plot for a given numerical feature  
@params : dataframe column mapped to an array, x-axis label of histogram, y-axis label of histogram, histogram title, box plot title 
@return : None 
'''
def plot_hist_and_box(array, xlabel_hist, ylabel_hist, title_hist, title_box):
    fig, (ax_hist, ax_box) = plt.subplots(1, 2, figsize=(12, 5))

    # Plot 1: Histogram visualizing frequency distribution of daily stocks prices and volume
    ax_hist.hist(array, bins=20)
    ax_hist.set_xlabel(xlabel_hist)
    ax_hist.set_ylabel(ylabel_hist)
    ax_hist.set_title(title_hist)

    # Plot 2: Boxplot visualizing the outliers among the dailty stock prices and volume 
    ax_box.boxplot(array, vert=False)
    ax_box.set_title(title_box)

    plt.show()

print("plot_hist_and_box() compiled ...")
```

```python
''' 
get_quartile_ranges() : a function to calculate and print the quartiles of a given dataframe column to analyze the box plot   
@params : pyspark dataframe, and the column for which we need to print the quartiles 
@return : None 
'''
def get_quartile_ranges(df, col_name):
    # calculate the quartile ranges 
    q1 = df.approxQuantile(col_name, [0.25], 0.001)[0]
    q3 = df.approxQuantile(col_name, [0.75], 0.001)[0]
    
    # Q2 is the median 
    q2 = df.approxQuantile(col_name, [0.50], 0.001)[0]
    
    # This is the Interquartile range where 50% values are there 
    IQR = q3 - q1
    
    # Use the 1.5IQR rule to calculate the lower and upper quartiles 
    
    q0 = (q1 - 1.5*IQR)
    q4 = (q1 + 1.5*IQR)
    
    
    print("First Quartile (Q1)  = ", q1)
    print("Third Quartile (Q3)  = ", q3)
    
    print("Median (Q2)  = ", q2)
    
    print("Interquartile Range (IQR)  = ", IQR)
    
    print("Lower Quartile (Q0)  = ", q0)
    print("Upper Quartile (Q4)  = ", q4)

print("get_quartile_ranges() compiled ...")
```

## Multivariate Analysis 

Calculate some aggregations on the categorical features, aggregated over either sum or mean of the numerical features. 

#### Yearly Stats   

```python
# Calculate yearly avg daily stock prices and total volumes 
yearly_stats = data_temp.groupBy("Year").agg(
    round(avg(col("Price High")), 2).alias("Avg_Price_High"),
    round(avg(col("Price Open")), 2).alias("Avg_Price_Open"),
    round(avg(col("Price Close")), 2).alias("Avg_Price_Close"),
    round(sum(col("Volume")), 2).alias("Total_Volume")
).sort('Year',ascending=False).na.drop()
yearly_stats.show()
```

#### Stats by Instrument (RIC)

```python
# Calculate the daily avg stock prices and total volume by the RICs 
instrument_stats = data_temp.groupBy("Instrument").agg(
    round(avg(col("Price High")), 2).alias("Avg_Price_High"),
    round(avg(col("Price Open")), 2).alias("Avg_Price_Open"),
    round(avg(col("Price Close")), 2).alias("Avg_Price_Close"),
    round(sum(col("Volume")), 2).alias("Total_Volume")
).na.drop()
instrument_stats.show()
```

#### Avg highest stock prices each Year by the instrument

```python
# An aggregation to calculate the avg highest stock price of each RIC per year 
yearly_instrument_stats = data_temp.groupBy("Year", "Instrument").agg(
    round(avg(col("Price High")), 2).alias("Avg_Price_High")).sort("Year","Avg_Price_High",ascending=False).na.drop()
yearly_instrument_stats.show()
```

### Analyzing trends in stock prices by the Instrument

#### Mapping the aggregated spark dataframe on Intsrument (RICs) into RDDs for faster performance

```python
# Avg daily high stock price by the RIC - spark DF to RDDs 
avg_PriceHigh_inst = instrument_stats.select("Instrument","Avg_Price_High").sort("Avg_Price_High", ascending=False).limit(10)
inst_priceHigh = avg_PriceHigh_inst.rdd.map(lambda row : row['Instrument']).collect()
priceHigh_array1 = avg_PriceHigh_inst.rdd.map(lambda row : row['Avg_Price_High']).collect()

# Avg daily opening stock price by the RIC - spark DF to RDDs
avg_PriceOpen_inst = instrument_stats.select("Instrument","Avg_Price_Open").sort("Avg_Price_Open", ascending=False).limit(10)
inst_priceOpen = avg_PriceOpen_inst.rdd.map(lambda row : row['Instrument']).collect()
priceOpen_array1 = avg_PriceOpen_inst.rdd.map(lambda row : row['Avg_Price_Open']).collect()

# Avg daily closing stock price by the RIC - spark DF to RDDs
avg_PriceClose_inst = instrument_stats.select("Instrument","Avg_Price_Close").sort("Avg_Price_Close", ascending=False).limit(10)
inst_priceClose = avg_PriceClose_inst.rdd.map(lambda row : row['Instrument']).collect()
priceClose_array1 = avg_PriceClose_inst.rdd.map(lambda row : row['Avg_Price_Close']).collect()

# Total daily stock volume by the RIC - spark DF to RDDs
total_volume_inst = instrument_stats.select("Instrument","Total_Volume").sort("Total_Volume", ascending=False).limit(10)
inst_volume = total_volume_inst.rdd.map(lambda row : row['Instrument']).collect()
volume_array1 = total_volume_inst.rdd.map(lambda row : row['Total_Volume']).collect()
volume_array1_fmt = [volume / 1e10 for volume in volume_array1]
```

### Correlation among features

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
from pyspark.mllib.stat import Statistics

# Convert to vector column first
df = data_temp.select("Price High","Price Open","Price Close", "Volume").na.drop()
cols = df.columns
assembler = VectorAssembler(inputCols=cols, outputCol="features")
df_assembled = assembler.transform(df)

corr_matrix = Correlation.corr(df_assembled, "features").collect()[0][0]
corr_values = corr_matrix.toArray()

scaled_corr_values = (corr_values + 1) / 2  # Scale between 0 and 1

# Create a DataFrame from the scaled correlation values
corr_df = pd.DataFrame(scaled_corr_values, columns=cols, index=cols)

corr_df
```

### Observations

- The target `Price High` is extremely closely correlated to the features `Price Close` and `Price Open`, almost 100%.
- This phenomenon is understandable, since they follow identical patterns as evident from the previous analysis. 
- The feature `Volume` does not exhibit the strong correlation with `Price High` and could be a good candiate in the feature set later on.

### Observations 

- In 2021, the average day highest stock price follows an incremental trend, with a significant spike from May to September.
- On the contrary, 2022 is quite the opposite of 2021 with a decremental trend right from January to December.
- It experiences a major upwards bump again in February 2023 before going on a decremental trend from April.
- 2020 experienced a very major volatility with sharp fall in March followed by a sudden sharp rise in April which persists until 2021 i.e. which is our year of spike.

### Recommendations 

Based on the Exploratory Data Analysis we have conducted, here are some of the recommendations for the next phase: 


1. **Extract-Transform-Load (ETL) - Data Cleansing**
 - Ensure that both the data files - main stocks dataset and the referential sectors dataset are in the same directory for feasible access. 
 - Ensure that the any indexed or unnamed columns from the CSV file are duly dropped or not read into the spark dataframes. This add constraints in checking for data duplication and makes the dataset more noisy.
 - Ensure that the NULL values are correctly imputed for the stock prices and volumes. Based on the univariate analysis, all the features are not normally distributed and are sensitive to outliers. So, we will not impute them with the mean.
 - For imputation, first drop the records where Date is NULL as they are irrelevant. Then, among the non-NULL Date records, impute the numerical features with the median. 
 - Ensure that all the features in the datasets are of the correct datatype.
 - Ensure to rename the column names of the features removing whitespaces, so that we can efficiently run SQL queries on the spark dataframes.
 - For relevant forecast, included the only from 2019 to 2023. There are some records with 2010 and 2018 that is noisy data that can negatively impact the accuracy of the forecasting models.



2. **Extract-Transform-Load (ETL) - Feature Engineering**
-  The open and close stock prices are nearly 100% correlated to highest price. So, they have a very strong impact of daily highest price.
-  But this highly strong correlation may lead to multicollinearity between the features and the target. So, this needs to be handled in the feature engineering accordingly. 
-  On the other hand, the correlation between volume and the highest stock price is significantly lower. So, it has a much lower impact. Its relation to the target will be explored in the feature engineering step. 
-  It is yet to be seen how does the feature importances play. In this step, analyze the feature importances and put together the insights from the correlation and feature importance calculation to build the feature vector.
